````python
"""
write a program to find the grades based on marks
1. if marks in between 80-100 - grade is A
2. if marks in between 60-79 - grade is B
3. if marks in between 40-59 - grade is C
4. if marks in between 10-39 - grade is D
5. if marks in between 9-0 - grade is F
6. if marks other than above output is - please enter correct marks
""" 
marks=int(input("please enter student marks here"))
if (marks>=80 and marks<=100):
    print(marks,"grade is A,congrats!!")
elif(marks>=60 and marks<=79):
    print(marks,"grade is B")
elif(marks>=40 and marks<=59):
    print (marks,"grade is C")
elif(marks>=10 and marks<=39):
    print(marks,"grade is D")
````

````python
# write a program to find the largest number from three numbers?
n1=int(input("enter first number here:"))
n2=int(input("enter second number here:"))
n3=int(input("enter third number here:"))

if n1 >= n2 and n1 >= n3:
    largest = n1
elif n2 >= n1 and n2 >= n3:
    largest = n2
else:
    largest = n3

print("The largest number is:", largest)
````

# iterations in python
1. iterations means loops to work with repeative tasks or repeative performances
2. python have two types of iterators
3. `for` loop
4. `while`loop

```python - for loop
for iter_variable in sequence:
    print("for is keyword to start the for loop")
    print("iter_variable is one type of variable to iterate data from sequence")
    print(" in is keyword to validate sequence is present or not")
    print("sequemce is collection like range,list,atring,tuple,...etc")

```python - while loop
while(condition):
    print("while is keyword to start the iterartion based on condition")
    print("condition is true thrn executed inside statements")
```


````python
#write a program to print 1 to 50 numbers using for loop?
for num in range(1,51):
    print (num)
    
````

````python
#write a program in iterate name using for loop
name="shivaji"
for char in name:
    print(char)
````

````python
#write a program to find the even numbers in between 100 to 150 using for loops?
for n in range (100,150):
    if (n%2==0):
        print(n,end=",")
````

````python
#write a program to find the vowels from sting using for loop?
name=str(input("please enter your name:"))
for char in name:
    if (char in "aeiouAEIOU"):
        print(char)
````

````python
#write a program to find the  non vowels from sting using for loop?
name=str(input("please enter your name:"))
for char in name:
    if (char not in "aeiouAEIOU"):
        print(char)
````

````python
#write a program to read name from list using for loop
names=["vikas","shiavji","aditya","chandra"]
for n in names:
    print("my name is",n)
````
